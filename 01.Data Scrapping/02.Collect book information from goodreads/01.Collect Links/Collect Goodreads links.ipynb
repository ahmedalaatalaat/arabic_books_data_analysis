```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument("--incognito")
chrome_options.add_argument("--nogpu")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--window-size=1280,1280")
chrome_options.add_argument("--enable-javascript")
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument('--disable-blink-features=AutomationControlled')
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
chrome_options.add_experimental_option('useAutomationExtension', False)
driver = webdriver.Chrome(options=chrome_options)


with open(f'good_reads_arabic_books_links.csv', 'w', encoding='utf-8-sig', newline="") as file:
    writer = csv.writer(file)
    writer.writerow(['Link'])

    for index in range(1, 26):
        print(f'Start Page # {index}')
        headers = {
            'User-Agent': random.choice(USER_AGENTS),
            "Accept-Encoding": "*",
            "Connection": "keep-alive"
        }
        url = f'https://www.goodreads.com/shelf/show/arabic?page={index}'
        driver.execute_cdp_cmd('Network.setUserAgentOverride', {"userAgent": random.choice(USER_AGENTS)})
        driver.get(url)

        soup = BeautifulSoup(driver.page_source, 'xml')
        books = soup.find_all('div', class_='elementList')
        for book in books:
            book_link = book.find('a').get('href')
            if 'show' in book_link:
                writer.writerow([f'https://www.goodreads.com{book_link}'])
```
